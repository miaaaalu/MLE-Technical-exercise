```python
# Third-party imports
import duckdb
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Scikit-learn imports
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest

# others 
import yaml
import warnings
warnings.filterwarnings('ignore')
pd.options.display.float_format = '{:.2f}'.format
```

```python
def load_and_describe_data(default_path, table_name, connection, file_path=None):
    if file_path is None:
        file_path = default_path
    
    data = pd.read_csv(file_path)
    connection.register(table_name, data)
    
    description = connection.execute(f"DESCRIBE {table_name}").fetchdf()
    print(f"{table_name.capitalize()} schema:")
    display(description)
    return data  # Return the loaded DataFrame

con = duckdb.connect()

# Set dataset paths
default_financial_path = 'dataset/business-financial-data-march-2024-csv.csv'
default_employment_path = 'dataset/machine-readable-business-employment-data-mar-2024-quarter.csv'

# Call the function and save the returned DataFrame
financial_data = load_and_describe_data(default_financial_path, 'financial_data', con)
employment_data = load_and_describe_data(default_employment_path, 'employment_data', con)
```
